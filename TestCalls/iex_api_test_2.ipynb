```python
def get_apicall_and_slicedata(stock_ticker):
    
    resp_data = requests.get(sandbox_url+'stock/'+stock_ticker+'/chart?token='+test_token)
    resp_json = pd.DataFrame(resp_data.json())  
    
    data = resp_json.set_index('date')
    returned = data[['close']]
    symbol = data.iloc[0,4]
    returned.rename(columns={'close':symbol},inplace=True)
    return returned
```

```python
aapl_df = get_apicall_and_slicedata("TSLA")
aapl_df.head()
```

```python
## GET /stock/{symbol}/chart/{range}/{date}
## chartCloseOnly

def get_apicall_and_slicedata_10years(stock_ticker):
    
    resp_data = requests.get(sandbox_url+'stock/'+stock_ticker+'/chart/10y?token='+test_token)
    resp_json = pd.DataFrame(resp_data.json())  
    
    data = resp_json.set_index('date')
    returned = data[['close']]
#     symbol = data.iloc[0,4]
    symbol = stock_ticker
    returned.rename(columns={'close':symbol},inplace=True)
    return returned
```
